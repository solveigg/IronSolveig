## Building function

```python
def give_me_audio_features (song):
    results = sp.search(song,limit=1,market="GB")
    song_id = results['tracks']['items'][0]['id']
    feat = sp.audio_features(results["tracks"]["items"][0]["id"])
    df = pd.json_normalize(feat)
    df = df.set_index('id')
    df = df.drop(['mode','key','type', 'uri', 'track_href', 
                            'analysis_url', 'duration_ms','time_signature'], axis=1)
    return df
```

```python
#user_input  = input('Enter a song: ',)

def give_me_audio_features_full (song):
    results = sp.search(song,limit=1,market="GB")
    #song_id = results['tracks']['items'][0]['id']
    feat = sp.audio_features(results["tracks"]["items"][0]["id"])
    df = pd.json_normalize(feat)
    df = df.set_index('id')
    df = df.drop(['mode','key','type', 'uri', 'track_href', 
                            'analysis_url', 'duration_ms','time_signature'], axis=1)
    df_scaled = scaler.transform(df)
    kmeans_number = kmeans.predict(df_scaled)[0]
    recom = X_scaled_df_clusters[X_scaled_df_clusters.cluster == kmeans_number]
    index_list = recom.index.values.tolist()    
    track_id = random.choice(index_list)   
    show_recom=  IFrame(src="https://open.spotify.com/embed/track/"+ track_id,
       width="320",
       height="80",
       frameborder="0",
       allowtransparency="true",
       allow="encrypted-media",
      )
    return show_recom
    
```
